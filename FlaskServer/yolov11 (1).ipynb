```python
import onnx
import onnx_tf
import tensorflow as tf

# Load the ONNX model
onnx_model = onnx.load("/content/runs/detect/train/weights/best.onnx")

# Convert ONNX model to TensorFlow SavedModel format
tf_rep = onnx_tf.backend.prepare(onnx_model)
tf_rep.export_graph("/content/runs/detect/train/weights/saved_model")  # Export to SavedModel format

# Convert TensorFlow SavedModel to TFLite
converter = tf.lite.TFLiteConverter.from_saved_model("/content/runs/detect/train/weights/saved_model")
tflite_model = converter.convert()

# Save the TFLite model
with open("/content/runs/detect/train/weights/best.tflite", "wb") as f:
    f.write(tflite_model)
```
